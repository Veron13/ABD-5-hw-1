```python
import pandas as pd
```

```python
data = pd.read_csv('продажи.csv')
data.head()
```

```python
data.columns = ['customer', 'Sum', 'Name', 'Data', 'Magazine']
data.head()
```

## Нарисовать pie-chart с прибылью по фильмам за все время

```python
data['Name'].unique()
```

```python
# Построем pie-график
import matplotlib
import matplotlib.pyplot as plt

data_group = data.groupby('Name').sum().reset_index()
sizes = data_group['Sum']
fig1, ax1 = plt.subplots(figsize=(8,8))
plt.title('pie-chart с прибылью по фильмам за все время')
ax1.pie(sizes, autopct='%1.1f%%',
        shadow=False, startangle=50)
ax1.axis('equal') 

plt.show()
```

## Получить top 5 самых продаваемых фильмов в каждом магазине

```python
data['Magazine'].unique()
```

```python
for x in data['Magazine'].unique():
    data_mag = data[data['Magazine'] == x]
    data_mag_5 = data_mag.groupby('Name').count().reset_index()
    print('top 5 самых продаваемых фильмов в '+ str(x))
    print(data_mag_5[['Name', 'Sum']].sort_values(by='Sum', ascending=False).head())
```
